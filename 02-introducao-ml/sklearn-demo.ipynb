# Treinando Modelos

```python
X = df.drop(['target', 'target_real_names'], axis=1)
X.head()
```

```python
y = df['target']
y.head()
```

## Regressão Logística

```python
from sklearn.linear_model import LogisticRegression

logreg_model = LogisticRegression(max_iter=1000)
logreg_model.fit(X, y)
```

```python
logreg_model.predict(X)
```

```python
logreg_model.predict_proba(X)[:10]
```

```python
iris_misteriosa = np.array([[2.3, 3.5, 1.1, 0.312]])
print(iris_misteriosa)
```

```python
print(logreg_model.predict(iris_misteriosa))
```

```python
print(logreg_model.predict_proba(iris_misteriosa))
```

## Árvore de Decisão

```python
from sklearn import tree

tree_model = tree.DecisionTreeClassifier()
tree_model.fit(X, y)
```

```python
print(tree_model.fit(X, y))
```

```python
df.loc[[1, 55, 102]]
```

```python
# prevendo labels
new_X = df.loc[[1, 55, 102]].drop(['target', 'target_real_names'], axis=1)
tree_model.predict(new_X)
```

```python
# prevendo probabilidades
tree_model.predict_proba(new_X)
```

```python
tree_model.predict(iris_misteriosa)
```

```python
tree_model.predict_proba(iris_misteriosa)
```

```python
# Plus: plotando a árvore
from sklearn.tree import plot_tree

fig, ax = plt.subplots(figsize=(14, 10))

plot_tree(tree_model, feature_names=X.columns.to_list(), class_names=['setosa', 'versicolor', 'virginica'], filled=True, rounded=True);
```

## Avaliando os modelos

Iremos avaliar os modelos usando a acurácia

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred_tree = tree_model.predict(X)
```

```python
accuracy_score(y, y_pred_tree)
```

```python
y_pred_logreg = logreg_model.predict(X)
```

```python
accuracy_score(y, y_pred_logreg)
```

A árvore de decisão apresentou uma melhor acurácia nos dados de treino.
