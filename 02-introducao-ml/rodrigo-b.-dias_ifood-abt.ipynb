## Criação da ABT

Será que variáveis como Recência, Frequência e Valor podem nos ajudar a prever se um dado cliente irá realizar um novo pedido no próximo mês?

Vamos criar variáveis que descrevem o comportamento do cliente com base em 1 mês e usar essas variáveis para prever se o cliente irá comprar no próximo mês (caso em que acontecerá um churn e talvez seja interessante oferecermos um cupom de desconto através de notificações no app).

**`Q5`** Crie um DataFrame chamado `df_features_train` que deverá conter os dados do período 2019-06-01 (inclusivo) até 2019-07-01 (exclusivo), ou seja deve conter somente os dados referente ao mês 6.

```python
df_features_train  = df_orders.query("order_timestamp_local >='2019-06-01'& order_timestamp_local < '2019-07-01'")
```

```python
df_features_train
```

**`Q6`** Crie um DataFrame chamado `df_target_train` que deverá conter os dados do período 2019-07-01 (inclusivo) até 2019-08-01 (exclusivo), ou seja deve conter somente os dados referente ao mês 7.

Depois selecione apenas a coluna `customer_id` e remova os duplicados.

```python
df_target_train = df_orders.query("order_timestamp_local >='2019-07-01' & order_timestamp_local < '2019-08-01'")
```

```python
#código para remover os clientes duplicados
df_target_train = df_target_train.filter(["customer_id"]).drop_duplicates()
df_target_train
```

**`Q7`** Faça o agrupamento no DataFrame `df_features_train` pela feature `customer_id`. Faça as agregações abaixo, renomeando o resultado para as seguintes colunas `uf`, `receita_1m`, `qtde_pedidos_1m` e `data_ultima_venda`, respectivamente.

- `customer_state_label` - `first`
- `paid_amount` - `sum`
- `order_id` - `nunique`
- `order_timestamp_local` - `max`

Lembre-se de utilizar o conceito que vimos em sala para renomear o resultado da agregação.

Também utilize o método `reset_index` para reinicar os indices.

Salve na variável `df_abt_train`.

```python
df_abt_train =(
    df_features_train
    .groupby("customer_id")
    .agg(uf = ('customer_state_label','first'),
         receita_1m = ("paid_amount","sum"),
         qtde_pedidos_1m = ("order_id","nunique"),
         data_ultima_venda =("order_timestamp_local" ,"max")
    )
    .reset_index()
)
```

```python
df_abt_train.head()
```

```python
df_abt_train.shape
```

**`Q8`** Utilize o método `.assign` do DataFrame para fazer as seguintes alterações:

- Altere o tipo da coluna `data_ultima_venda` para o formato de data utilizando o comando `pd.to_datetime` com o parâmetro `utc=True`.
- Crie uma nova coluna chamada `data_ref` com o valor de `2019-07-01` e faça a converção utilizando o comando `pd.to_datetime` com o parâmetro `utc=True`
- Crie uma nova coluna chamada `recencia` e salve os dias referentes a subtração da coluna `data_ref` e `data_ultima_venda`. Lembre-se de extrair os dias (`.dt.days`)
- Por fim, selecione apenas as seguintes colunas: `['data_ref', 'customer_id', 'receita_1m', 'qtd_pedidos_1m', 'recencia']`

O resultado dessas alterações devem ser salvas na variável `df_abt_train`


```python
df_abt_train = df_abt_train.assign(
    data_ultima_venda=pd.to_datetime(df_abt_train["data_ultima_venda"], utc=True),
    data_ref=pd.to_datetime("2019-07-01", utc=True),
    recencia=lambda x: (x["data_ref"] - x["data_ultima_venda"]).dt.days
).drop(["uf"], axis=1)

df_abt_train
```

**`Q9`** Faça o `merge` com o DataFrame `df_abt_train` com o `df_target_train` com os seguintes parametros:

- `how='left'`
- `on='customer_id'`
- `indicator=True`

Salve o merge na variável `df_abt_train`.


```python
df_abt_train = df_abt_train.merge(df_target_train, on = "customer_id", how = "left", indicator = True)
```

```python
df_abt_train
```

**`Q12`** Crie uma ABT Out of Time avançando 1 mês no código anterior. Porque devemos criar uma ABT Out of Time?

Utilize o mês 08 para criar essa ABT.

Porque devemos criar uma ABT Out of Time?

Resposta: A ABT Out of Time permite simular o desempenho do modelo, replicando condições de produção. Isso é essencial para verificar sua capacidade de generalizar para cenários futuros.


```python
df_oot_train  = df_orders.query("order_timestamp_local >='2019-07-01' & order_timestamp_local < '2019-08-01'")
```

```python
df_target_oot = df_orders.query("order_timestamp_local >='2019-08-01' & order_timestamp_local < '2019-09-01'").filter(["customer_id"]).drop_duplicates()
df_target_oot
```

```python
df_abt_oot =(
    df_oot_train
    .groupby("customer_id")
    .agg(uf = ('customer_state_label','first'),
         receita_1m = ("paid_amount","sum"),
         qtde_pedidos_1m = ("order_id","nunique"),
         data_ultima_venda =("order_timestamp_local" ,"max")
    )
    .reset_index()
)
df_abt_oot
```

```python
df_abt_oot = df_abt_oot.assign(
    data_ultima_venda=pd.to_datetime(df_abt_oot["data_ultima_venda"], utc=True),
    data_ref=pd.to_datetime("2019-08-01", utc=True),
    recencia=lambda x: (x["data_ref"] - x["data_ultima_venda"]).dt.days
).drop(["data_ultima_venda"], axis = 1)

df_abt_oot
```

```python
df_abt_oot = df_abt_oot.merge(df_target_oot, on = "customer_id", how = "left", indicator = True)
df_abt_oot = df_abt_oot.assign(churn_next_month = lambda df: np.where(df["_merge"]=="left_only",1,0)).drop("_merge",axis=1)
df_abt_oot
```

```python
df_abt_oot.value_counts("churn_next_month", normalize=True)
```
