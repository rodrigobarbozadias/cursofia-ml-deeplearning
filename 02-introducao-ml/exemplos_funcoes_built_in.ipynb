## 1. dict()

```
# Cria um dicionário
meu_dicionario = dict(nome="João", idade=30, cidade="São Paulo")
print(meu_dicionario)
```

## 2. enumerate()

```
# Retorna um objeto enumerado, útil para iterar com índices
frutas = ["maçã", "banana", "laranja"]
for indice, fruta in enumerate(frutas):
    print(indice, fruta)
```

## 3. float()

```
# Converte um número ou string em ponto flutuante
numero = float(10)
print(numero)
```

## 4. help()

```
# Mostra a documentação de um objeto ou função
help(print)
```

## 5. input()

```
# Recebe uma entrada do usuário
nome = input("Digite seu nome: ")
print(f"Olá, {nome}!")
```

## 6. int()

```
# Converte um número ou string em um inteiro
numero = int("20")
print(numero)
```

## 7. len()

```
# Retorna o tamanho de uma sequência (string, lista, etc.)
frutas = ["maçã", "banana", "laranja"]
print(len(frutas))
```

## 8. list()

```
# Converte um objeto iterável em uma lista
tupla = (1, 2, 3)
lista = list(tupla)
print(lista)
```

## 9. max()

```
# Retorna o maior valor em uma sequência
numeros = [10, 20, 30, 40]
print(max(numeros))
```

## 10. min()

```
# Retorna o menor valor em uma sequência
numeros = [10, 20, 30, 40]
print(min(numeros))
```

## 11. print()

```
# Imprime uma saída na tela
print("Olá, Mundo!")
```

## 12. range()

```
# Gera uma sequência de números
for i in range(5):
    print(i)
```

## 13. reversed()

```
# Inverte uma sequência
lista = [1, 2, 3, 4]
print(list(reversed(lista)))
```

## 14. round()

```
# Arredonda um número para o número de casas decimais fornecido
numero = 3.14159
print(round(numero, 2))
```

## 15. set()

```
# Cria um conjunto (set), que não permite valores duplicados
conjunto = set([1, 2, 2, 3, 4])
print(conjunto)
```

## 16. str()

```
# Converte um objeto em string
numero = 123
texto = str(numero)
print(texto)
```

## 17. sum()

```
# Retorna a soma de todos os elementos em um iterável
numeros = [1, 2, 3, 4]
print(sum(numeros))
```

## 18. type()

```
# Retorna o tipo de um objeto
print(type(10))
print(type("Olá"))
print(type([1, 2, 3]))
```
