```python
from ydata_profiling import ProfileReport
```

```python
import pandas as pd
```

```python
pasta_raiz = '/content/drive/MyDrive/datasets'
```

```python
df_limite_clientes = pd.read_csv(f'{pasta_raiz}/limite_clientes.csv', sep=';')
df_limite_clientes.head(3)
```

```python
df_estados_br = pd.read_csv(f'{pasta_raiz}/estados_br.csv', sep=';')
df_estados_br.head(3)
```

```python
df_final = df_limite_clientes.merge(df_estados_br, how='inner', left_on='uf', right_on='uf_sigla')
df_final = df_final.rename(columns={'uf_x': 'uf', 'uf_y': 'uf_nome'})
df_final = df_final.drop(columns='uf')
df_final.head()
```

```python
profile = ProfileReport(df_final, title='Limites de clientes')
```

```python
# Utilize o comando abaixo para visualizar o relatório que foi gerado.
# Pode levar alguns minutos para gerar o relatório.
profile
```

```python
# Caso não seja possível visualizar o relatório aqui, salve em HTML
# Pode levar alguns minutos para gerar o relatório.
profile.to_file(f'{pasta_raiz}/relatorio_conjunto_dados_limite_clientes.html')
```
