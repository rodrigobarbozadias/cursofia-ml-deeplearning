# Questão 2: Comparação entre dois valores

Escreva um código que compara dois números e imprime qual é o maior.


```python
a=15
b=10

if (a>b):
  print(f"a={a} é maior que b={b}")
elif(b>a):
  print("b={b} é maior que a={a}a")
else:
    print("a={a} é igual a b={b}")
```

# Questão 3: Uso de operadores lógicos

Escreva um código que verifica se um número está entre 10 e 20 (inclusive) usando operadores lógicos.

```python
x = 15

if (x >= 10) and (x <= 20):
    print("o numero está entre 10 e 20.")
else:
    print("o numero não está entre 10 e 20")
```

# Questão 4: Avaliação de múltiplas condições

Crie um código que verifique se um número é positivo, negativo ou zero e imprima a respectiva mensagem.

```python
numero = 5

if (numero>0):
  print(f"o numero {numero} é positivo")
elif(numero<0):
  print(f"o numero {numero} é negativo")
else:
    print(f"o numero {numero} é zero")
```
